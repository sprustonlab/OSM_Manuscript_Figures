```python
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import igraph

import numpy as np
```

```python
fig, axs = plt.subplots(1,4, figsize = (15,4))
for i in [1,2,3,4]:##range(1, 4):
    A = np.load('../FILES/A_stage_'+ str(i)+'.npy')
    edge_color = np.load('../FILES/edge_color_stage_'+ str(i)+'.npy')

    g = igraph.Graph.Adjacency((A > 0).tolist())

    igraph.plot(g,vertex_color= edge_color, target= axs[i-1])
fig.savefig('../FIGURES/transition_graphs.pdf', format='pdf', dpi=500)
```
